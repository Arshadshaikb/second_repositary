```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
for cols in ['species', 'island', 'sex']:
    sns.boxplot(x='bill_length_mm',
                y=cols,
                hue='species',
                data=data)
    plt.show()
```

```python
sns.scatterplot(data, x='bill_depth_mm', y='body_mass_g', hue='species')
```
